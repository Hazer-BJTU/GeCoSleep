```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'font.family': 'Times New Roman',          # 字体（优先使用无衬线字体）
    'font.size': 18,                 # 基础字号
    'axes.titlesize': 24,            # 标题字号
    'axes.labelsize': 24,            # 坐标轴标签字号
    'xtick.labelsize': 20,           # X轴刻度字号
    'ytick.labelsize': 20,           # Y轴刻度字号
    'xtick.major.pad': 15,  # X轴刻度标签与轴的距离（默认：3.5）
    'ytick.major.pad': 10,  # Y轴刻度标签与轴的距离（默认：3.5）
    'axes.grid': False,               # 显示网格
    'grid.alpha': 1,               # 网格透明度
    'axes.prop_cycle': plt.cycler(color=plt.cm.tab10.colors)  # 使用高对比度色环
})

# 设置数据
tasks = ['Well trained', 'After task 2', 'After task 3', 'After task 4']
methods = ['GeCoSleep', 'DER 512', 'DT2W', 'Lwf', 'ER 512', 'A-GEM 512', 'TA-A-GEM 512', 'Fine-Tuning', 'EWC', 'None']

# 生成随机数据作为示例
data1 = np.array(
    [
        [78.2, 78.8, 78.9, 78.9],
        [78.7, 78.3, 77.8, 77.6],
        [78.6, 78.7, 76.8, 76.1],
        [78.1, 78.2, 76.2, 75.4],
        [78.3, 77.1, 76.7, 75.8],
        [78.3, 76.2, 75.1, 74.7],
        [78.1, 76.0, 73.9, 71.9],
        [78.8, 73.6, 72.3, 71.6],
        [78.1, 73.4, 71.8, 70.0],
        [78.6, 72.5, 69.1, 69.1]
    ]
)

data2 = np.array(
    [
        [75.8, 75.0, 74.7, 74.3],
        [76.4, 74.4, 73.3, 72.9],
        [76.1, 73.9, 70.3, 69.9],
        [75.5, 72.9, 69.5, 68.7],
        [75.8, 72.4, 71.7, 70.8],
        [75.6, 71.3, 70.3, 69.7],
        [75.5, 71.2, 67.8, 65.7],
        [76.2, 66.2, 64.9, 65.3],
        [75.5, 66.3, 64.2, 63.5],
        [75.8, 65.0, 61.0, 62.8]
    ]
)

data3 = np.array(
    [
        [0.000, 0.006, 0.018, 0.015],
        [0.000, -0.004, -0.006, -0.024],
        [0.000, 0.001, -0.008, -0.038],
        [0.000, 0.001, -0.014, -0.049],
        [0.000, -0.012, -0.03, -0.073],
        [0.000, -0.021, -0.05, -0.113],
        [0.000, -0.021, -0.055, -0.163],
        [0.000, -0.052, -0.084, -0.214],
        [0.000, -0.047, -0.082, -0.216],
        [0.000, -0.061, -0.119, -0.246]
    ]
)

data4 = np.array(
    [
        [0.000, -0.008, -0.008, -0.018],
        [0.000, -0.02, -0.027, -0.054],
        [0.000, -0.022, -0.061, -0.1],
        [0.000, -0.026, -0.064, -0.111],
        [0.000, -0.034, -0.051, -0.107],
        [0.000, -0.043, -0.064, -0.141],
        [0.000, -0.043, -0.09, -0.208],
        [0.000, -0.1, -0.133, -0.269],
        [0.000, -0.092, -0.134, -0.272],
        [0.000, -0.108, -0.176, -0.305]
    ]
)

data3 *= 100
data4 *= 100

# 创建图形和子图
line_styles = ['-', '--', '-.', ':']
markers = ['o', 's', '^', 'v', 'D', 'P', '*', 'X']
colors = [
    '#1F77B4',  # 蓝
    '#FF7F0E',  # 橙
    '#2CA02C',  # 绿
    '#D62728',  # 红
    '#9467BD',  # 紫
    '#8C564B',  # 棕
    '#E377C2',  # 粉
    '#7F7F7F',  # 灰
    '#A1C9F4',  # 浅蓝
    '#FFB482',  # 肉粉
    '#8DE5A1',  # 薄荷
    '#FF9F9B',  # 珊瑚
    '#D0BBFF',  # 薰衣草
    '#DEBB9B',  # 卡其
    '#FAB0E4',  # 粉紫
    '#CFCFCF'  # 浅灰
]

# 创建图形
plt.figure(figsize=(40, 9))

# 第一幅图
plt.subplot(1, 4, 1)
for i, method in enumerate(methods):
    style_idx = i % len(line_styles)
    marker_idx = i % len(markers)
    plt.plot(tasks, data1[i],
             linestyle=line_styles[style_idx],
             marker=markers[marker_idx],
             color=colors[i],
             label=method)
#plt.title('Forgetting of the first task (Acc)', pad=20)
plt.xlabel('A. Forgetting of the first task (Acc)', labelpad=15)
plt.ylabel('Accuracy', labelpad=10)
plt.ylim(68, 80)
plt.grid(True, alpha=0.3)
#plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0, framealpha=1)

# 第二幅图
plt.subplot(1, 4, 2)
for i, method in enumerate(methods):
    style_idx = i % len(line_styles)
    marker_idx = i % len(markers)
    plt.plot(tasks, data2[i],
             linestyle=line_styles[style_idx],
             marker=markers[marker_idx],
             color=colors[i],
             label=method)
#plt.title('Forgetting of the first task (mF1)', pad=20)
plt.xlabel('B. Forgetting of the first task (mF1)', labelpad=15)
plt.ylabel('macro F1 Score', labelpad=10)
plt.ylim(60, 77)
plt.grid(True, alpha=0.3)
#plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0, framealpha=1)

# 第三幅图
plt.subplot(1, 4, 3)
for i, method in enumerate(methods):
    style_idx = i % len(line_styles)
    marker_idx = i % len(markers)
    plt.plot(tasks, data3[i],
             linestyle=line_styles[style_idx],
             marker=markers[marker_idx],
             color=colors[i],
             label=method)
#plt.title('Forgetting of the first task (mF1)', pad=20)
plt.xlabel('B. Forgetting of the first task (mF1)', labelpad=15)
plt.ylabel('macro F1 Score', labelpad=10)
plt.ylim(-25, 2)
plt.grid(True, alpha=0.3)
#plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0, framealpha=1)

# 第四幅图
plt.subplot(1, 4, 4)
for i, method in enumerate(methods):
    style_idx = i % len(line_styles)
    marker_idx = i % len(markers)
    plt.plot(tasks, data4[i],
             linestyle=line_styles[style_idx],
             marker=markers[marker_idx],
             color=colors[i],
             label=method)
#plt.title('Forgetting of the first task (mF1)', pad=20)
plt.xlabel('B. Forgetting of the first task (mF1)', labelpad=15)
plt.ylabel('macro F1 Score', labelpad=10)
plt.ylim(-31, 2)
plt.grid(True, alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0, framealpha=1)

plt.tight_layout()
plt.savefig('./figure/forgetting_comparison.pdf')
plt.show()
```
